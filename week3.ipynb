Scraping method:Use pandas to read the table into a pandas dataframe



```python
import pandas as pd
d=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df_neighbor=d[0]
df_neighbor.head()
```

Code to Ignore cells with a borough that is Not assigned

```python
df_neighbor=df_neighbor.set_index('Borough')
df_neighbor=df_neighbor.drop("Not assigned", axis=0)
```

```python
df1=df_neighbor.reset_index()

df1.head()
```

```python
#swapped column names
df2=df1[['Postcode','Borough','Neighbourhood']]
df2.head()
```

Group rows into unique row with postcode and neighborhoods separated with a comma

```python
df3=df2.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
df3.head()
```

Replacing Not assigned neighborhood with corresponding Borough values

```python



df3['Neighbourhood'][df3['Neighbourhood']=='Not assigned']=df3['Borough']
```

```python
#final dataframe
df3
```

```python
df3.shape
```

Renamed column to Postal code

```python
df4=df3.rename(columns={'Postcode':'Postal Code'})
df4.head()
```

Read csv file with geospatial data to new dataframe-(starts 2nd submission)

```python
df5=pd.read_csv('http://cocl.us/Geospatial_data')
df5
```

Merge previous dataframe with new dataframe to form dataframe with all deatils

```python
df6=pd.merge(left=df4,right=df5,left_on='Postal Code',right_on='Postal Code')
df6
```

Explore and cluster neighbourhoods of Toronto(starts 3rd submission)

select Borough that contains word Toronto

```python
df7 =  df6[df6['Borough'].str.contains('Toronto')].reset_index(drop=True)
df7
```

```python
#removed postalcode column for making analysis simple
df8=df7.drop(['Postal Code'],axis=1)
df8.head()
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(df8['Borough'].unique()),
        df8.shape[0]
    )
)
```

Use geopy library to get the latitude and longitude values of Toronto

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

Create a map of Toronto with neighborhoods superimposed on top

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map

for lat, lng, borough, neighbourhood in zip(df8['Latitude'], df8['Longitude'], df8['Borough'], df8['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto) 
    #print(lat)
    
map_toronto
```

Apply k-means clustering to group the neighbourhoods

```python
df9=df8.drop(['Borough','Neighbourhood'],axis=1)
df9.head()
```

```python
kclusters=4
kmeans = KMeans(init="k-means++", n_clusters=4, n_init=12).fit(df9)
kmeans.labels_[0:10]
```

Add cluster labels to cleaned dataframe

```python
#create a copy of df8 
local_df = df8.copy(deep=True)
local_df.head()
```

```python
# insert cluster labels to new df
local_df.insert(0,"Cluster labels",kmeans.labels_)
local_df.head()
```

create dataframes for each cluster

```python
dataframe_cluster0=local_df.loc[local_df['Cluster labels']==0]
dataframe_cluster1=local_df.loc[local_df['Cluster labels']==1]
dataframe_cluster2=local_df.loc[local_df['Cluster labels']==2]
dataframe_cluster3=local_df.loc[local_df['Cluster labels']==3]



```

```python
#display shape each cluster
print('Cluster 0',dataframe_cluster0.shape)
print('Cluster 1',dataframe_cluster1.shape)
print('Cluster 2',dataframe_cluster2.shape)
print('Cluster 3',dataframe_cluster3.shape)
```
